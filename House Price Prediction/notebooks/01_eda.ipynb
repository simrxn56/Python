```python
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(bins=30, figsize=(15, 10))
plt.suptitle('Distribution of Numeric Features')
plt.savefig('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/outputs/notebook 1/numeric_distribution.png')
plt.show()
plt.close()
```

```python
corcol = df.drop('city', axis=1).corr()['price'].sort_values(ascending=False)
corcol
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='sqft_living', y='price', line_kws={'color': 'red'})
plt.title('Price vs Sqft_living')
plt.savefig('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/outputs/notebook 1/scatterplots/Price vs sqftliving.png')
plt.show()
plt.close()
```
