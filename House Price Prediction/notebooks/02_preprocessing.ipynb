```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/data/cleaned/data.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
cat_cols = df.select_dtypes(include='object').columns.tolist()
```

```python
df = pd.get_dummies(df, cat_cols, drop_first=False)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=11)
```

```python
train_X.to_csv('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/data/processed/train_X.csv', index=False)
test_X.to_csv('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/data/processed/test_X.csv', index=False)
train_y.to_csv('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/data/processed/train_y.csv', index=False)
test_y.to_csv('C:/Users/LENOVO/OneDrive/Desktop/PYTHON/House Price Prediction/data/processed/test_y.csv', index=False)
```
